```python
import pandas as pd
```

```python
dat = pd.read_csv('hand_labeled_tops_324.csv')
```

```python
dat.head()
```

```python
## build vocab
dat['combined_tax_att'] = ''
for index,row in dat.iterrows():
    tax_cat = row['taxonomy_cat']
    pattern = row['pattern']
    color = row['color']
    
    combined_cat = tax_cat+'_'+str(pattern)+'_'+str(color)
    combined_cat = combined_cat.replace('[','')
    combined_cat = combined_cat.replace(']','')
    combined_cat = combined_cat.replace("'",'')
    
    dat.loc[index,'combined_tax_att'] = combined_cat
    print(combined_cat)
```

```python
dat.head()
```

```python
def convert_agent_output_to_judge_input(input_str):
    broken_out = input_str.replace('_',' ')
    input_str = broken_out +' '+input_str
    return input_str
convert_agent_output_to_judge_input('top_melange_grey')
```

```python
dat['for_judge'] = ''
for index, row in dat.iterrows():
    tax_att = row['combined_tax_att']
    dat.loc[index,'for_judge'] = convert_agent_output_to_judge_input(tax_att)
```

```python
corpus = dat['for_judge'].tolist()
```

```python
vectorizer = CountVectorizer()
vectorizer = vectorizer.fit(corpus)
print(vectorizer.get_feature_names())
```

```python
test_outfit = ['top top top yellow solid top_yellow_solid top green melange top_green_melange top pink solid top_pink_solid']

vectorized_otfit = vectorizer.transform(test_outfit)
vectorized_otfit.toarray()
```

```python
len(vectorizer.get_feature_names())
```
